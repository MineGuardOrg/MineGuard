```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    roc_curve,
    auc
)


df = pd.read_csv("dataset_ml_final.csv")

# Variables para el modelo
X = df[["value", "type_cod"]]
y = df["label"]
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("\n REPORTE DE CLASIFICACIN:\n")
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Matriz de Confusi贸n - Riesgo vs Normal")
plt.show()
```

```python

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curva ROC - Detecci贸n de Riesgo")
plt.legend(loc="lower right")
plt.show()

```

```python

df["label"].value_counts().plot(kind="bar")
plt.title("Distribuci贸n de Clases")
plt.xlabel("Clase (0 = Normal | 1 = Riesgo)")
plt.ylabel("Cantidad de Registros")
plt.show()
```

```python
nuevo_dato = np.array([[520, 7]])

pred = model.predict(nuevo_dato)[0]
prob = model.predict_proba(nuevo_dato)[0][1]

print("\n PREDICCIN NUEVA:")
print("Valor:", nuevo_dato[0][0])
print("Tipo codificado:", nuevo_dato[0][1])
print("Predicci贸n:", "RIESGO" if pred == 1 else "NORMAL")
print("Probabilidad de riesgo:", round(prob * 100, 2), "%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


df = pd.read_csv("dataset_ml_final.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])

X = df[["value", "type_cod"]]
y = df["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
df["prob_riesgo"] = model.predict_proba(X)[:, 1]
```

```python
sensores = df["type"].unique()

for sensor in sensores:

    df_s = df[df["type"] == sensor].sort_values("timestamp")

    plt.figure(figsize=(10, 5))

    plt.plot(df_s["timestamp"], df_s["value"], label="Valor real", linewidth=2)
    plt.plot(df_s["timestamp"], df_s["prob_riesgo"], label="Probabilidad de Riesgo", linewidth=2)

    plt.title(f"Sensor: {sensor} | Riesgo a trav茅s del tiempo")
    plt.xlabel("Tiempo")
    plt.ylabel("Valor / Probabilidad")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    plt.show()  
```

```python
df_caidas = df[df["type"].isin(["ax", "ay", "az"])]

prob_caida_user = df_caidas.groupby("user_id")["prob_riesgo"].mean()

plt.figure(figsize=(8, 5))
prob_caida_user.plot(kind="bar")
plt.title("Probabilidad Promedio de Ca铆da por Usuario")
plt.xlabel("Usuario")
plt.ylabel("Probabilidad de Ca铆da")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("dataset_ml_final.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python

for sensor in df["type"].unique():

    df_s = df[df["type"] == sensor].sort_values("timestamp")

    plt.figure()
    plt.plot(df_s["timestamp"], df_s["value"])
    plt.title(f"Serie de Tiempo - Sensor {sensor}")
    plt.xlabel("Tiempo")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()
```

```python

for sensor in df["type"].unique():

    df_s = df[df["type"] == sensor]

    plt.figure()
    plt.scatter(df_s["timestamp"], df_s["value"])
    plt.title(f"Dispersi贸n - Sensor {sensor}")
    plt.xlabel("Tiempo")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()
```

```python

for sensor in df["type"].unique():

    df_s = df[df["type"] == sensor]

    plt.figure()
    plt.boxplot(df_s["value"])
    plt.title(f"Distribuci贸n de Valores - Sensor {sensor}")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()
```

```python
for sensor in df["type"].unique():

    df_s = df[df["type"] == sensor]
    conteo = df_s["label"].value_counts()

    plt.figure()
    plt.pie(conteo)
    plt.title(f"Proporci贸n Riesgo vs Normal - Sensor {sensor}")
    plt.show()
```

```python
for sensor in df["type"].unique():

    df_s = df[(df["type"] == sensor) & (df["label"] == 1)]

    plt.figure()
    plt.scatter(df_s["timestamp"], df_s["value"])
    plt.title(f"Eventos de Riesgo en el Tiempo - Sensor {sensor}")
    plt.xlabel("Tiempo")
    plt.ylabel("Valor")
    plt.grid(True)
    plt.show()
```

```python

riesgo_usuario = df.groupby("user_id")["label"].mean()

plt.figure()
plt.plot(riesgo_usuario.index, riesgo_usuario.values, marker='o')
plt.title("Promedio de Riesgo por Usuario")
plt.xlabel("Usuario")
plt.ylabel("Probabilidad de Riesgo")
plt.grid(True)
plt.show()
```
